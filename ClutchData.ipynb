```python
from boxscore_utils import create_boxscore_one_game

canada_games = df[df.Nationality=='CAN'].rsc.unique()
total_clutch_can = pd.DataFrame()
for rsc in canada_games:
    game = df[(df.rsc==rsc)&(df.Nationality == 'CAN')&(df.period=='Q4')&(df.game_mins < 6)&(abs(df.SA-df.SB) < 6)]
    total_clutch_can = pd.concat([total_clutch_can,game])
# total_clutch = create_boxscore_one_game(total_clutch_can)
```

```python
from boxscore_utils import get_stats_from_actions, add_columns

# Filter relevant columns for analysis
player_stats_df = total_clutch_can[['rsc','name', 'Nationality', 'Action']]

# Group by player and perform aggregation
player_stats = player_stats_df.groupby(['rsc','name', 'Nationality', 'Action']).size().reset_index(name='count')
games_per_player = player_stats_df.groupby(['name', 'Nationality']).rsc.nunique()

boxscore_df = get_stats_from_actions(player_stats)
boxscore_df = boxscore_df.groupby(['Player', 'Nationality']).sum().drop(columns='rsc')
boxscore_df_per_game = boxscore_df.divide(games_per_player[boxscore_df.reset_index()['Player']], axis=0)
final_boxscore = add_columns(boxscore_df_per_game)
```

```python
final_boxscore.sort_values(by='PER', ascending=False)
```

```python
from boxscore_utils import create_boxscore_one_game

no_canada = df[df.Nationality!='CAN'].rsc.unique()
total_clutch_can = pd.DataFrame()
for rsc in no_canada:
    game = df[(df.rsc==rsc)&(df.Nationality != 'CAN')&(df.period=='Q4')&(df.game_mins < 6)&(abs(df.SA-df.SB) < 6)]
    total_clutch_can = pd.concat([total_clutch_can,game])
# total_clutch = create_boxscore_one_game(total_clutch_can)
from boxscore_utils import get_stats_from_actions, add_columns

# Filter relevant columns for analysis
player_stats_df = total_clutch_can[['rsc','name', 'Nationality','Action']]

# Group by player and perform aggregation
player_stats = player_stats_df.groupby(['rsc','name', 'Nationality','Action']).size().reset_index(name='count')
games_per_player = player_stats_df.groupby(['name', 'Nationality']).rsc.nunique()

boxscore_df = get_stats_from_actions(player_stats)
boxscore_df = boxscore_df.groupby(['Player', 'Nationality']).sum().drop(columns='rsc')
boxscore_df_per_game = boxscore_df.divide(games_per_player[boxscore_df.reset_index()['Player']], axis=0)
final_boxscore = add_columns(boxscore_df_per_game)
```
