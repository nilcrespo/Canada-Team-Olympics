```python
import pandas as pd
all_greece = pd.DataFrame()
filenames = ['121112-CRO-GRE.csv', '121092-DOM-GRE.csv', '121090-GRE-EGY.csv', '121111-GRE-SLO.csv']
for file_name in filenames:
    df = pd.read_csv(file_name)
    df['game_code'] = file_name.split('-')[0]
    all_greece = pd.concat([all_greece, df])
all_greece.to_csv('all_greece.csv', index=False)
```
