```python
from rotations_utils import get_dict_players_numbers, get_heatmap_dicts_df, compute_new_columns_and_dfs, all_plots_game
import pandas as pd

def process_files(filenames, team_, team_acronim):
    cumulative_minutes = {}
    cumulative_heatmaps = [None] * 4
    cumulative_plusminus = pd.DataFrame()
    cumulative_cumulative_pm = pd.DataFrame()
    all_country = pd.DataFrame()

    for i, file_name in enumerate(filenames):
        file_name = file_name.replace('.csv', '')
        if file_name.split('-')[1] == team_acronim:
            team_str = 'A'
            team = file_name.split('-')[1]
        else:
            team_str = 'B'
            team = file_name.split('-')[2]
        teams = file_name.split('-')[1:3]
        print(teams)

        all_pbp = pd.read_csv(file_name + '.csv')
        all_pbp['game_code'] = file_name.split('-')[0]
        all_country = pd.concat([all_country, all_pbp])

        dict_players_numbers = get_dict_players_numbers(all_pbp, team)

        time_percentages_dfs = get_heatmap_dicts_df(all_pbp, team_str, team, dict_players_numbers)
        all_pbp_no_end_q, heatmap_plus_minus = compute_new_columns_and_dfs(all_pbp, team_str)

        minutes = []
        for key in dict_players_numbers.keys():
            minutes.append(all_pbp[all_pbp[f'X_{int(key)}_j_{team_str}'] == 1]['time_elapsed'].sum())
        player_minutes = dict(zip(dict_players_numbers.values(), minutes))

        # Aggregate player minutes
        for key, value in player_minutes.items():
            if key in cumulative_minutes:
                cumulative_minutes[key] += value
            else:
                cumulative_minutes[key] = value

        # Aggregate heatmaps
        for j in range(4):
            if cumulative_heatmaps[j] is None:
                cumulative_heatmaps[j] = time_percentages_dfs[j]
            else:
                cumulative_heatmaps[j] = cumulative_heatmaps[j].add(time_percentages_dfs[j], fill_value=0)

        # Aggregate plusminus
        cumulative_plusminus = pd.concat([cumulative_plusminus, heatmap_plus_minus])

        # Aggregate cumulative_pm
        cumulative_pm = all_pbp_no_end_q.groupby(['quarter', 'minute'])['+/-'].quantile([0, 0.25, 0.5, 0.75, 1]).reset_index()
        cumulative_cumulative_pm = pd.concat([cumulative_cumulative_pm, cumulative_pm])

    all_country.to_csv(f'all_{team_.lower()}.csv', index=False)


    # Finalize aggregated results
    sorted_sum_dict = dict(sorted(cumulative_minutes.items(), key=lambda item: item[1], reverse=True))

    total_percentages = []
    for j in range(4):
        total_percentages.append(cumulative_heatmaps[j] / len(filenames))
        total_percentages[j] = total_percentages[j].reindex(list(sorted_sum_dict.keys()))

    average_plusminus = cumulative_plusminus.groupby(['quarter', 'minute'])['plusminus'].mean().reset_index()
    average_cumulative_pm = cumulative_cumulative_pm.groupby(['quarter', 'minute', 'level_2'])['+/-'].mean().reset_index()

    all_plots_game(total_percentages[0], total_percentages[1], total_percentages[2], total_percentages[3], average_plusminus, average_cumulative_pm, team_, None)

# Example usage
filenames = ['117647-FRA-BRA.csv', '117648-JPN-FRA.csv', '117650-FRA-GER.csv']
team_ = 'France'
team_acronim = 'FRA'
process_files(filenames, team_, team_acronim)
```
